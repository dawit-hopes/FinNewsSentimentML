```python
import pandas as pd
from  sentiment import SentimentAnalyzer 
```

```python
df = pd.read_csv("../data/newsData/raw_analyst_ratings.csv")
```

```python
df['date'] = pd.to_datetime(df['date'], format='ISO8601')
df['date'] = df['date'].dt.date
df = df.set_index('date')
```

```python
analyzer = SentimentAnalyzer()
tickers = ['AMZN','META','MSFT','NVDA','AAPL','GOOG']

for ticker in tickers: 
    correlation_coefficient = analyzer.perform_analysis(ticker, df)
    print(f"{ticker} Correlation Coefficient (r): {correlation_coefficient}")
```
